### Variables Description

**id**: This is a unique identifier for each listing on Airbnb. It's used internally by Airbnb to distinguish between different listings.

**name:** This column contains the name or title of the listing. It provides a brief description of the accommodation.

**host_id:** Similar to the 'id' column, this is a unique identifier for each host who has listed their property on Airbnb. It allows Airbnb to associate listings with specific hosts.

**host_name:** This column contains the name of the host who owns or manages the listing. It identifies the person responsible for the property.

**neighbourhood_group:** This column indicates the larger geographical area or district in which the listing is located. It might represent neighborhoods grouped together for analysis or presentation purposes.

**neighbourhood:** This column specifies the specific neighborhood or locality where the listing is situated. It provides more granular information about the location of the accommodation.

**room_type:** This column describes the type of accommodation being offered, such as 'Entire home/apt', 'Private room', or 'Shared room'. It indicates whether the guest will have exclusive access to the entire property or will be sharing it with others.

**price:** This column represents the nightly price (in the local currency) charged for renting the accommodation. It gives an indication of the cost of staying at the listing.

**minimum_nights:** This column specifies the minimum number of nights a guest must book in order to stay at the listing. It indicates the host's policy regarding the duration of stays.

**number_of_reviews:** This column indicates the total number of reviews that the listing has received from previous guests. It provides an indication of the popularity and possibly the quality of the accommodation.

**calculated_host_listings_count:** This column shows the total number of listings managed by the host. It provides insight into the scale of the host's operation on Airbnb.

**availability_365:** This column represents the number of days in a year for which the listing is available for booking. It provides information about the listing's availability throughout the year.

#### **Chart 1-** **Which type of room/home is most preffered by the guests?**

````python
# Chart - 1 visualization code
room_types=airbnb1.groupby(['room_type'])['number_of_reviews'].max().reset_index()
room_types=room_types.sort_values(by='number_of_reviews',ascending=False)
room_types
````

````python
# Visualise the Data
# Plot the pie chart with explode and shadow
plt.figure(figsize=(8, 6))
plt.pie(room_types['number_of_reviews'], labels=room_types['room_type'],
        colors=['orange', 'green', 'blue'],
        autopct='%1.2f%%',
        startangle=60,
        explode=[0.1, 0, 0],
        shadow=True, textprops={'fontsize': 14})

# Set the title
plt.title('Type of room/home most preferred by the guests', color='red', fontsize=20)

# Customize the legend
legend = plt.legend(room_types['room_type'], title='Room Types', loc='center left', bbox_to_anchor=(1, 0.5, 0.5, 0.5))
plt.setp(legend.get_title(), color='red')  # Change legend title color
legend.get_frame().set_facecolor('cyan')  # Change legend background color

# Display the pie chart
plt.show()
````

**Insight:-** **Private room** is the most preferred room type with **40.04%** of the total Reviews, indicating that **Private room** type are the most popular or commonly reviewed type accommodation among the options listed.

##### 1. Why did you pick the specific chart?

This pie chart give you a clear and easy-to-interpret visual representation of the number of bookings by room type. and shows in percentage the analysis.

##### 2. What is/are the insight(s) found from the chart?

**Private room** is the most preferred room type with **40.04%** of the total Reviews, indicating that **Private room** type are the most popular or commonly reviewed type accommodation among the options listed.
